## Popularity Based Recommender

```python
rating_with_name=rating.merge(books,on='ISBN')
```

```python
rating_with_name
```

```python
#num_rating is number of people who gave ratings.
num_rating_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
```

```python
avg_rating_df=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
```

```python
popular_df= num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

```python
popular_df
```

## collaborative based recommendation system

```python
# we are only going to choose the users who have rated 200 books.
# and books that is rated by minimum 50 person 
```

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating']>200
criteria1=x[x].index
```

```python
filtered_rating=rating_with_name[rating_with_name['User-ID'].isin(criteria1)]


y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books=y[y].index
```

```python
final_rating=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates().shape
```

```python
pt=final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items =sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
   
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('Dark Matter')
```
